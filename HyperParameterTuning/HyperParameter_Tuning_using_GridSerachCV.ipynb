```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier 
from sklearn.linear_model import LogisticRegression 
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn import datasets
digits = datasets.load_digits()
```

```python
digits.feature_names
```

```python
model_parameters={
    'svm':{
          "model":SVC(gamma ='auto'),
          "parameters": {
            'C' : [1,5,10,15,20,25],
            'kernel' :  ['linear', 'poly', 'rbf', 'sigmoid']
        }
    },

    'RandomForest':{
        "model": RandomForestClassifier(),
        "parameters": {
            "n_estimators" : [20,40,50,60,80,100],
            "criterion" : ["gini", "entropy"]
        }
    },

    'Logistic_Regression':{
        "model": LogisticRegression(),
        "parameters":{
            'C' : [0.01,0.05,0.1,1,5,10],
            "solver" : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
        }
    },

    'Gaussian':{
        "model": GaussianNB(),
        "parameters":{}
    },

    'Multinomial':{
        "model": MultinomialNB(),
        "parameters":{
            'alpha': [0.01,0.1,0.5,1.0,2.5,5.0,10.0]
        }
    },

    'Decision_Tree':{
        "model": DecisionTreeClassifier(),
        "parameters":{
            'criterion' : ["gini", "entropy"],
            'splitter' : ["best", "random"]
        }
    }

}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
scores = []
```

```python
for model_name,mp in model_parameters.items():
  clf = GridSearchCV(mp['model'], mp['parameters'], cv=5, return_train_score = False)
  clf.fit(digits.data, digits.target)
  scores.append({
      'model' : model_name,
      'best_score' : clf.best_score_,
      'best_param' : clf.best_params_
  })
  
```

```python
df = pd.DataFrame(scores)
```

###The Clear winner is SVC() at an accuracy of 96.8% with C=1 and kernel='poly'
