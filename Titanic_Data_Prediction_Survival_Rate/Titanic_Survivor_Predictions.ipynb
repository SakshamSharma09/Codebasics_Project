```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv("/content/Titanic_Survivors.csv")
```

```python
titanic_data.head()
```

```python
titanic_data.isnull().sum()
```

```python
drp_col = titanic_data.isnull().sum()[titanic_data.isnull().sum()>(0.35 * titanic_data.shape[0])]
#It provides the column name which has more than 35% null values.
drp_col
```

```python
titanic_data.drop(drp_col.index, axis=1,inplace=True )
titanic_data.head()
```

```python
titanic_data["Family Size"] = titanic_data['SibSp'] + titanic_data['Parch']
titanic_data.head()
```

```python
titanic_data.drop(['SibSp','Parch'], axis=1, inplace=True)
```

```python
titanic_data.fillna(titanic_data.mean(), inplace=True)
```

```python
titanic_data['Embarked'].describe()
```

```python
titanic_data['Embarked'].fillna('S', inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.drop(['Name','Ticket'],axis='columns',inplace=True)
```

```python
titanic_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
le_embarked = LabelEncoder()
```

```python
titanic_data['sex'] = le_sex.fit_transform(titanic_data['Sex'])
titanic_data['embarked'] = le_embarked.fit_transform(titanic_data['Embarked'])
```

```python
titanic_data.head()
```

```python
titanic_data.drop(['Sex','Embarked'], axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.drop(['PassengerId'], axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
X = titanic_data.drop('Survived',axis=1)
y = titanic_data.Survived
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
